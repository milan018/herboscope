```python
#Importing Libraries
import os
import numpy as np

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import load_img, ImageDataGenerator
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten
import json
```

```python
# âœ… Count Images
count = 0
base_dir = 'Medicinal Leaf Dataset/'
dirs = os.listdir(base_dir)
for dir in dirs:
    files = os.listdir(os.path.join(base_dir, dir))
    print(f'{dir} Folder has {len(files)} Images')
    count += len(files)
print(f'Images Folder has {count} Images')
```

```python
base_dir = 'Medicinal Leaf Dataset'
img_size = 224
batch_size = 32

train_ds = tf.keras.utils.image_dataset_from_directory(
    base_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(img_size, img_size),
    batch_size=batch_size
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    base_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(img_size, img_size),
    batch_size=batch_size
)

plant_names = train_ds.class_names
with open("class_names.json", "w") as f:
    json.dump(plant_names, f)
```

```python
from sklearn.utils import class_weight
# Extract all labels
all_labels = []
for _, labels in train_ds:
    all_labels.extend(labels.numpy())

class_weights_dict = dict(enumerate(class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(all_labels),
    y=all_labels
)))

```

```python
AUTOTUNE = tf.data.AUTOTUNE

data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.03),
    layers.RandomZoom(0.05),
    layers.RandomContrast(0.05),
])

# Apply augmentation only to training set
augmented_train_ds = train_ds.map(
    lambda x, y: (data_augmentation(x, training=True), y),
    num_parallel_calls=AUTOTUNE
)

# Prefetch
augmented_train_ds = augmented_train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(img_size, img_size, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False  # Phase 1

model = models.Sequential([
    layers.Rescaling(1./255),
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.3),
    layers.Dense(len(plant_names), activation='softmax')  # âœ… softmax added
])
```

```python

model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),  # âœ… updated
    metrics=['accuracy']
)

history = model.fit(
    augmented_train_ds,
    validation_data=val_ds,
    epochs=15,
    class_weight=class_weights_dict
)
```

```python
base_model.trainable = True
fine_tune_at = 100  # Freeze first 100 layers
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)

history_fine = model.fit(
    augmented_train_ds,
    validation_data=val_ds,
    epochs=20,  # â¬…ï¸ Longer fine-tuning
    class_weight=class_weights_dict
)
```

```python
def Classify_images(image_path):
    input_image = tf.keras.utils.load_img(image_path, target_size=(img_size, img_size))
    input_image_array = tf.keras.utils.img_to_array(input_image)  # âš ï¸ no manual /255
    input_image_exp_dim = tf.expand_dims(input_image_array, 0)

    predictions = model.predict(input_image_exp_dim, verbose=0)
    predicted_probs = predictions[0]
    predicted_index = np.argmax(predicted_probs)
    
    print("\nðŸ” Top Predictions:")
    top_5 = np.argsort(predicted_probs)[-5:][::-1]
    for i in top_5:
        print(f"{plant_names[i]}: {predicted_probs[i]*100:.2f}%")
        
    return plant_names[predicted_index]
```

```python
# âœ… Classification Function

img_size = 224  # Must match model input
with open("class_names.json") as f:
    plant_names = json.load(f)
def classify_images(image_path):
    input_image = tf.keras.utils.load_img(image_path, target_size=(img_size, img_size))
    input_image_array = tf.keras.utils.img_to_array(input_image) / 255.0
    input_image_exp_dim = tf.expand_dims(input_image_array, 0)

    predictions = model(input_image_exp_dim, training=False)  # âœ… Disable augmentation & dropout
    result = tf.nn.softmax(predictions[0])
    predicted_class = plant_names[np.argmax(result)]
    return predicted_class

```

```python
def predict_image(model, image_path, class_names, img_size=180):
    img = tf.keras.utils.load_img(image_path, target_size=(img_size, img_size))
    img_array = tf.keras.utils.img_to_array(img)
    img_array = img_array / 255.0  # âœ… same normalization as training
    img_batch = tf.expand_dims(img_array, 0)  # âœ… add batch dimension

    predictions = model.predict(img_batch)
    probs = tf.nn.softmax(predictions[0])  # âœ… convert logits to probabilities

    predicted_index = tf.argmax(probs).numpy()
    predicted_class = class_names[predicted_index]
    confidence = probs[predicted_index].numpy()

    return predicted_class, confidence
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

true_labels = []
pred_labels = []

for images, labels in val_ds:
    predictions = model.predict(images)
    predicted_classes = tf.argmax(predictions, axis=1)

    true_labels.extend(labels.numpy())
    pred_labels.extend(predicted_classes.numpy())

cm = confusion_matrix(true_labels, pred_labels)
plt.figure(figsize=(8, 8))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=plant_names, yticklabels=plant_names, cmap="YlGnBu")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(true_labels, pred_labels, target_names=plant_names))
```
